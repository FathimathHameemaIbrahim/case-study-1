# 1. Read the dataset to the python environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv('cars_data.csv')
```

```python
data1
```

# 2. Check for the null values present in the dataset

```python
#Gives True for null values
data1.isna()
```

```python
#To confirm the presence of null values in a series
data1.isnull().values.any()
```

# 3. Plot a bar graph of male vs female buyers participated in the sales

```python
#To get Count of Female and male buyers
data1['Buyer Gender'].value_counts()
```

```python
#Plotting the Bar graph
data1['Buyer Gender'].value_counts().plot(kind='bar',
                                         figsize=(8,8),
                                         title=("male vs female buyers participated in the sales"),
                                         color=('blue','red'))
plt.xlabel('Gender')
plt.ylabel('Count in person')
```

# 4. Find the top 5 cars based on their sales price

```python
#Importing dataset into DataFrame
df = pd.DataFrame(data1)
df
```

```python
#slicing and displaying columns 'Make' and 'Sale Price'
df1 = df.loc[:,['Make','Sale Price']]
df1
```

```python
#To display the top 5 cars based on their sales price 
df1.nlargest(5, 'Sale Price')
```

# 5. Find the least 5 cars based on their Resell price

```python
#slicing and displaying columns 'Make' and 'Resell Price'
df2 = df.loc[:,['Make','Resell Price']]
df2
```

```python
#To display the least 5 cars based on their Resell Price
df2.nsmallest(5, 'Resell Price')
```
